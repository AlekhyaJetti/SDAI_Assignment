```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Load the Wine dataset
wine=pd.read_csv('WineQT.csv')
```

```python
wine.head()
```

```python
X = wine[['fixed acidity', 'volatile acidity']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Decide on the number of clusters, K
K = 3

# Create an instance of KMeans with K clusters and fit the data
kmeans = KMeans(n_clusters=K, random_state=42)
kmeans.fit(X_scaled)
```

```python
# Get the cluster assignments for each data point
labels = kmeans.labels_

# Get the cluster centroids
centroids = kmeans.cluster_centers_
```

```python

# Plot the clusters with different colors using the specified columns
plt.scatter(X_scaled[labels == 0][:, 0], X_scaled[labels == 0][:, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X_scaled[labels == 1][:, 0], X_scaled[labels == 1][:, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X_scaled[labels == 2][:, 0], X_scaled[labels == 2][:, 1], s = 100, c = 'green', label = 'Cluster 3')

# Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], s = 100, c = 'yellow', label = 'Centroids')
plt.legend()
plt.xlabel('Column 1')
plt.ylabel('Column 2')
plt.title('K-means Clustering for Custom Wine Dataset')

plt.show()
```
